```python
x = {"Adelie": None, "Chinstrap": None, "Gentoo": None}
y = {"Adelie": None, "Chinstrap": None, "Gentoo": None}
```

```python
u = {"Adelie": None, "Chinstrap": None, "Gentoo": None}
v = {"Adelie": None, "Chinstrap": None, "Gentoo": None}
```

```python
penguins = {}
```

```python
penguins["Adelie"] = penguinsSet[penguinsSet['species'] == 'Adelie']
penguins["Chinstrap"] = penguinsSet[penguinsSet['species'] == 'Chinstrap']
penguins["Gentoo"] = penguinsSet[penguinsSet['species'] == 'Gentoo']
```

```python
[penguins["Adelie"].shape, penguins["Chinstrap"].shape, penguins["Gentoo"].shape, penguinsSet.shape]
```

```python
x['Adelie'] = penguins["Adelie"]["bill_depth_mm"].to_numpy()
y['Adelie'] = penguins["Adelie"]["bill_length_mm"].to_numpy()
x['Chinstrap'] = penguins["Chinstrap"]["bill_depth_mm"].to_numpy()
y['Chinstrap'] = penguins["Chinstrap"]["bill_length_mm"].to_numpy()
x['Gentoo'] = penguins["Gentoo"]["bill_depth_mm"].to_numpy()
y['Gentoo'] = penguins["Gentoo"]["bill_length_mm"].to_numpy()
```

```python
u['Adelie'] = penguins["Adelie"]["flipper_length_mm"].to_numpy()
v['Adelie'] = penguins["Adelie"]["body_mass_g"].to_numpy()
u['Chinstrap'] = penguins["Chinstrap"]["flipper_length_mm"].to_numpy()
v['Chinstrap'] = penguins["Chinstrap"]["body_mass_g"].to_numpy()
u['Gentoo'] = penguins["Gentoo"]["flipper_length_mm"].to_numpy()
v['Gentoo'] = penguins["Gentoo"]["body_mass_g"].to_numpy()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
s = fig.get_size_inches()
s[0] = s[0] * 2
fig.set_size_inches(s)

ax1.set_title("bill depth vs length (mm)")
ax1.set_xlabel("bill_depth_mm")
ax1.set_ylabel("bill_length_mm")
ax1.scatter(x['Adelie'], y['Adelie'], c='green')
ax1.scatter(x['Chinstrap'], y['Chinstrap'], c='red')
ax1.scatter(x['Gentoo'], y['Gentoo'], c='orange')

ax2.set_title("flipper length (mm) vs body mass (g)")
ax2.set_xlabel("flipper_length_mm")
ax2.set_ylabel("body_mass_g")
ax2.scatter(u['Adelie'], v['Adelie'], c='green')
ax2.scatter(u['Chinstrap'], v['Chinstrap'], c='red')
ax2.scatter(u['Gentoo'], v['Gentoo'], c='orange')
```
